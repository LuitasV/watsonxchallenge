```python
import matplotlib.pyplot as plt


updated_data = results
models = [entry['model'] for entry in updated_data]
metrics_keys = set(key for entry in updated_data for key in entry['metrics'])
metrics_values = {metric: [] for metric in metrics_keys}

for entry in updated_data:
    for metric in metrics_keys:
        metrics_values[metric].append(entry['metrics'].get(metric, 0))

# Plotting the updated data
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.15  # width of the bars
n = len(models)
indices = range(len(metrics_keys))  # the x locations for the groups

for i, model in enumerate(models):
    ax.bar([x + i * width for x in indices], [updated_data[i]['metrics'][metric] for metric in sorted(metrics_keys)], width, label=model)

ax.set_xlabel('Métricas')
ax.set_ylabel('Puntuación')
ax.set_title('Comparativa de modelos: watsonx challenge')
ax.set_xticks([x + (n-1) * width / 2 for x in indices])
ax.set_xticklabels(sorted(metrics_keys))
ax.legend()

plt.show()
```
